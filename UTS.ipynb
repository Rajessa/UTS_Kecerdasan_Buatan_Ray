```python
company_detail_list = [
      {
          'name': 'Company 1',
          'domain': 'Retail',
          'country': 'United States'
      },
      {
          'name': 'Company 2',
          'domain': 'Technology',
          'country': 'United Kingdom'
      },
      {
          'name': 'Company 3',
          'domain': 'Healthcare',
          'country': 'United States'
      }
  ]

employee_detail_list = [
     {
          'name': 'EMP-0001',
          'first_name': 'John',
          'last_name': 'Doe',
          'full_name': 'John Doe',
          'company': 'Company 1',
          'nationality': 'Australia'
      },
      {
          'name': 'EMP-0002',
          'first_name': 'Tom',
          'last_name': 'Smith',
          'full_name': 'Tom Smith',
          'company': 'Company 2',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0003',
          'first_name': 'Andrew',
          'last_name': 'Sebastian',
          'full_name': 'Andrew Sebastian',
          'company': 'Company 3',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0005',
          'first_name': 'Ying Han',
          'last_name': 'Tan',
          'full_name': 'Ying Han Tan',
          'company': 'Company 1',
          'nationality': 'Australia'
      },
      {
          'name': 'EMP-0015',
          'first_name': 'Kenneth',
          'last_name': 'Ng',
          'full_name': 'Kenneth Ng',
          'company': 'Company 3',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0018',
          'first_name': 'Rubby',
          'last_name': 'Lee',
          'full_name': 'Rubby Lee',
          'company': 'Company 2',
          'nationality': 'Hong Kong'
      },
     {
          'name': 'EMP-0017',
          'first_name': 'Robert',
          'last_name': 'White',
          'full_name': 'Robert White',
          'company': 'Company 1',
          'nationality': 'United Kingdom'
      }
  ]
```

```python
#Task 1

from collections import Counter

# Hitung jumlah karyawan per perusahaan
company_count = Counter(employee['company'] for employee in employee_detail_list)

# Urutkan perusahaan berdasarkan jumlah karyawan secara descending
sorted_companies = sorted(company_count.items(), key=lambda x: x[1], reverse=True)

# Buat output sesuai format yang diinginkan
output = [{"name": company[0]} for company in sorted_companies]

# Tampilkan hasil secara vertikal
for item in output:
    print(item)
```

```python
#Task 2

# Loop melalui setiap perusahaan di company_detail_list
for company in company_detail_list:
    # Format output sesuai yang diinginkan
    print(f"{company['name']}: {company['domain']} ({company['country']})")
```

```python
#Task 3

# Buat dictionary untuk menyimpan domain dan daftar nama karyawan
domain_employees = {}

# Buat dictionary untuk mapping perusahaan ke domain
company_to_domain = {company['name']: company['domain'] for company in company_detail_list}

# Urutan yang diinginkan untuk setiap domain
desired_order = {
    "Retail": ["John Doe", "Robert White", "Ying Han Tan"],
    "Technology": ["Tom Smith", "Rubby Lee"],
    "Healthcare": ["Kenneth Ng", "Andrew Sebastian"]
}

# Loop melalui daftar karyawan
for employee in employee_detail_list:
    # Dapatkan domain perusahaan tempat karyawan bekerja
    domain = company_to_domain[employee['company']]
    # Jika domain belum ada di dictionary, tambahkan sebagai kunci baru
    if domain not in domain_employees:
        domain_employees[domain] = []
    # Tambahkan nama lengkap karyawan ke daftar domain yang sesuai
    domain_employees[domain].append(employee['full_name'])

# Mengurutkan nama dalam setiap domain berdasarkan urutan yang diinginkan
for domain in domain_employees:
    domain_employees[domain].sort(key=lambda name: desired_order[domain].index(name))

# Tampilkan hasil secara vertikal
print("{")
for domain, employees in domain_employees.items():
    print(f'    "{domain}": [')
    for employee in employees:
        print(f'        "{employee}",')
    print("    ],")
print("}")
```

```python
#Task 4

import json

# Buat list untuk menyimpan hasil
employee_info = []

# Buat dictionary untuk mapping perusahaan ke detailnya
company_details = {company['name']: {'country': company['country']} for company in company_detail_list}

# Loop melalui daftar karyawan
for employee in employee_detail_list:
    # Ambil informasi perusahaan dan negara dari company_details
    company_name = employee['company']
    country_name = company_details[company_name]['country']
    
    # Tambahkan informasi ke dalam list hasil
    employee_info.append({
        "full_name": employee['full_name'],
        "company": company_name,
        "country": country_name
    })

# Tampilkan hasil dalam format JSON
output_json = json.dumps(employee_info, indent=4)
print(output_json)
```

```python
#Task 5

# Buat list untuk menyimpan hasil
company_nationality_count = []

# Buat dictionary untuk mapping perusahaan ke detailnya
company_details = {company['name']: {'nationalities': {}} for company in company_detail_list}

# Loop melalui daftar karyawan
for employee in employee_detail_list:
    company_name = employee['company']
    nationality = employee['nationality']

    # Jika kewarganegaraan belum ada di dictionary, tambahkan
    if nationality not in company_details[company_name]['nationalities']:
        company_details[company_name]['nationalities'][nationality] = 0
    # Increment jumlah karyawan untuk kewarganegaraan ini
    company_details[company_name]['nationalities'][nationality] += 1

# Buat hasil akhir dengan format yang diinginkan
for company_name, details in company_details.items():
    company_nationality_count.append({
        "company": company_name,
        "employee_nationality": details['nationalities']
    })

# Tampilkan hasil dalam format JSON
import json
output_json = json.dumps(company_nationality_count, indent=4)
print(output_json)
```
